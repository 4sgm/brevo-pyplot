```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
categories = ['Delivered', 'Estimated Openers', 'Trackable Openers',
              'Unique Clickers', 'Hard Bounced', 'Soft Bounced', 'Bounced',
              'Complaint', 'Blocked', 'Unsubscribe', 'Double Opt-In']

batch_data = {
    'Brevo': ['Batch1-A_B-599999-539901-03-21', 'Batch1-C-541103-03/22', 'Batch1-D-541105-03/22-3PM', 'Batch1-E-LQR-03-252'],
    'Delivered': [88, 85, 82],
    'Estimated Openers': [13, 16, 8],
    'Trackable Openers': [11, 11, 6],
    'Unique Clickers': [8, 9, 6],
    'Hard Bounced': [13, 14, 20],
    'Soft Bounced': [2, 4, 1],
    'Bounced': [14, 18, 21],
    'Complaint': [0, 0, 0],
    'Blocked': [0, 0, 0],
    'Unsubscribe': [2, 2, 0],
    'Double Opt-In': [5, 8, 4]
}

# Creating DataFrame
df = pd.DataFrame(batch_data)
df.set_index('Brevo', inplace=True)

# Plotting
ax = df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Email Metrics by Batch')
plt.xlabel('Batch')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
