```python
import matplotlib.pyplot as plt

# Data
categories = ['Sent', 'Delivered', 'Estimated Openers', 'Trackable Openers', 'Unique Clickers', 'Hard Bounced',
              'Soft Bounced', 'Bounced', 'Complaint', 'Blocked', 'Unsubscribe', 'Double Opt-In']

batch1_data = [1343, 1181, 168, 150, 104, 168, 22, 190, 3, 0, 24, 64]
batch2_data = [258, 220, 42, 29, 24, 36, 10, 46, 0, 0, 4, 20]
batch3_data = [319, 262, 24, 19, 20, 65, 2, 67, 0, 0, 1, 14]

# Plotting
x = range(len(categories))
width = 0.25

fig, ax = plt.subplots()
bar1 = ax.bar(x, batch1_data, width, label='Batch1-A_B-599999-539901-03-21')
bar2 = ax.bar([i + width for i in x], batch2_data, width, label='Batch1-C-541103-03/22')
bar3 = ax.bar([i + width*2 for i in x], batch3_data, width, label='Batch1-D-541105-03/22-3PM')

ax.set_xlabel('Categories')
ax.set_ylabel('Counts')
ax.set_title('Email Metrics by Batch')
ax.set_xticks([i + width for i in x])
ax.set_xticklabels(categories, rotation=90)
ax.legend()

plt.tight_layout()
plt.show()
```
